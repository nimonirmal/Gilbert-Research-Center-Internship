```python
#Import necessary library
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import quad
```

```python
# Define the function
def z(t):
  return t
def y(t):
  return np.sqrt(8)* np.sin(t+np.pi/2)
def x(t):
  return np.sqrt(8)*(np.sqrt(1-np.cos(2*t)/np.sqrt(2)))
```

```python
# Generating data points
t_val= np.linspace(0,3,1000)     # interval of t= 0<t<3
z_val=z(t_val)
y_val=y(t_val)
x_val=x(t_val)
```

```python
# Plot the 3D graph
fig= plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection ='3d')
ax.plot( x_val, y_val, z_val, color= 'red', linewidth=2)

ax.set_title('3-D line plot representation of Single-stranded RNA SARS-CoV-2 Covid-19 Structure')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
plt.show()
```

# Calculate the length of RNA SARS-CoV-2 structure from  0<t<3

```python
import sympy as smp
from scipy.misc import derivative
import sympy as smp

# find out the derivative of x(t)
t = smp.symbols('t')
x = 2 * smp.sqrt((1 - smp.cos(2*t)))
dx_dt = diff(x, t)  # diff() = used to find out the derivative
print("Derivative of x is dx/dt = ",dx_dt)
```

```python
# find out the derivative of y(t)
y=smp.sqrt(8)*smp.cos(t)
dy_dt = diff(y, t)
print("Derivative of y is dy/dt = ",dy_dt)
```

```python
# find out the derivative of z(t)
z=t
dz_dt = diff(z, t)
print("Derivative of z is dz/dt = ",dz_dt)
```

```python
# Add the Squares of derivatives and take the squareroot
dl= smp.sqrt(dx_dt**2 + dy_dt**2 + dz_dt**2).simplify()
dl
```

```python
#Integrate the 'dl' with respect to t from interval 0<t<3
Length =smp.integrate(dl,(t,0,3))
print("Length of RNA Structure in interval from 0<t<3 is : ", Length)
```
