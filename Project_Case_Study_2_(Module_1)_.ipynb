```python
df= pd.read_csv(r'cyclone.csv')
df
```

```python
# Create new dataframe with desired columns
new_df = df[['Year','Cyclonic Disturbances - TOTAL']]
new_df
```

```python
#Information about the data
new_df.info()
```

```python
# Generate Sample data greater than 30
sample_data=new_df.sample(n=50, random_state=1,ignore_index = True )
sample_data
```

```python
#Calculate the Sample mean and Standard Deviation
n=len(sample_data['Cyclonic Disturbances - TOTAL'])
sample_mean = sample_data['Cyclonic Disturbances - TOTAL'].mean()
sample_std = sample_data['Cyclonic Disturbances - TOTAL'].std()
print("Sample Size:", n)
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)

#Calculate Population Mean i.e mu
mu=new_df['Cyclonic Disturbances - TOTAL'].mean()
print("Population Mean:", mu)
```

**State the Hypothesis**            
Null Hypothesis (HO) : The mean cyclonic disturbance total is equal to the population mean (i.e 12.31).      
Alternative Hypothesis(Ha): The mean cyclonic disturbance total is greater than the population mean (i.e 12.31).      

```python
#Calculate Z-Statistics
Z=(sample_mean-mu)/(sample_std/np.sqrt(n))
print("Z-Statistics:", Z)

#Compare Z-Statistics with critical value
critical_value=1.645
print("Critical Value:", critical_value)

if Z>critical_value:
  print("Reject the Null Hypothesis HO")
else:
  print("Fail to Reject the Null Hypothesis HO")

#Compute P-value for one tail test
P_value = 1-stats.norm.cdf(Z)
print("P-value:", P_value)
```

```python
#Visualization
plt.figure(figsize=(10,6))
plt.plot(df['Year'],df['Cyclonic Disturbances - TOTAL'],label='Cyclonic Disturbance Total',linestyle='-')
plt.axhline(y=sample_mean, color='r', linestyle='-', label='Sample Mean')
plt.axhline(y=mu, color='g', linestyle='--', label='Population Mean')
plt.xlabel('Year')
plt.ylabel('Cyclonic Disturbance Total')
plt.title('Cyclonic Disturbance Total Over Years')
plt.legend()
plt.grid(True)
plt.show()
```
