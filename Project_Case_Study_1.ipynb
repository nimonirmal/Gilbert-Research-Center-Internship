```python
import numpy as np
A = np.array ([ [1,1,0],
                [1,0,0],
                [0,0,1],
               [0,1,-1] ])
print("Size of matrix A is :",A.shape)
print("\n Matrix A is given by :\n", A)

B=np.array([50,10,10,30])
print("\n Size of matrix B is :",B.shape)
print("\n Matrix B is given by :\n", B)
```

```python
#Finding unknown solution point of Matrix AX=B form
X=np.linalg.pinv(A).dot(B)
print(f"Solution is given by : X1 = {np.round(X[0])}, X2 = {np.round(X[1])}, X3 = {np.round(X[2])}")
```

```python
#Norm of a Solution Matrix
print("Norm of Unkown Varaible X is : ",np.linalg.norm(X))
print("-------------------------------------------------------")
#Rank of a Matrix
print("Rank of A is :", np.linalg.matrix_rank(A))
print("-------------------------------------------------------")
#Rank of Augmented Matrix
rank_A_B = np.linalg.matrix_rank(np.hstack((A,B.reshape(-1,1))))
print("Rank of Augmented Matrix i.e rank(A/B) is :",rank_A_B)
print("-------------------------------------------------------")
#Rank Nulity of Matrix
rank_nullity=(A.shape[1]-np.linalg.matrix_rank(A))
print(f"Rank Nulity of Matrix A is :{rank_nullity}")
```
