**Module 1                           A.T = A**           
Verify the Symmetric Matrix Property  

```python
#Import necessary library
import numpy as np
```

```python
# First Method
def check_symmetric(A):
  B=A.T                       # .T()= Transpose
  if np.array_equal(A,B):     # .array_equal() = to compare the matrix
    print("A is a symmetric matrix")
  else:
    print("A is not a symmetric matrix")

A=np.array([ [1,4,-3],
              [4,1,7],
              [-3,7,0]  ])

# Calling Function
check_symmetric(A)
```

```python
# Second Method
def is_symmetric(matrix):
  m, n= len(matrix), len(matrix[0])
  if m!=n:
    print("Not a Symmetric Matrix")

  for i in range(m):
    for j in range(n):
      if matrix[i][j] != matrix[j][i]:
         return False

  return True


matrix=np.array([ [1,4,-3],
                  [4,1,7],
                  [-3,7,0]  ])

if is_symmetric(matrix):
  print("It is a Symmetric Matrix")
else:
  print("It is not a Symmetric Matrix")
```

**Module 2: A.T = (A)^-1**           
Verify the Orthagonal Matrix Property

```python
import numpy as np

def is_orthagonal(matrix):
  matrix_transpose = matrix.transpose()      # for transpose of matrix
  matrix_inverse = np.linalg.inv(matrix)     # for inverse of matrix
  print("The matrix is :\n",matrix)
  if np.array_equal(matrix_transpose, matrix_inverse):
    print("Given Matrix is Orthagonal matrix")
  else:
    print("Given Matrix is not Orthagonal matrix")

matrix=np.array([[1,0,0],
                 [0,1,0],
                 [0,0,1]])
is_orthagonal(matrix)
```

**Module 4- I^(-1) = I (or) I^(-1) != I**               
Where I is the Identity Matrix                        
Verify the above property which is true.

```python
import numpy as np
#Creating an identity matrix of 3*3
I=np.eye(3)
print("Identity Matrix is :\n",I)
I_inverse = np.linalg.inv(I)
print("Inverse matrix of Identity Matrix is:\n",I_inverse)
if np.array_equal(I,I_inverse):
  print("\n Inverse of a Identity Matrix is equal to Identity Matrix")
else:
  print("\n Inverse of a Identity Matrix is not equal to Identity Matrix")
```

**Module 6- v1 = (3,5,4), v2 = (3,-5,4), v3= (4,0,-3)**               
Find the ortogonal and orthonormal for above vector?                        
Find the angle between each vector?

```python
import numpy as np
# Given Vectors are:
v1 = np.array([3,5,4])
v2 = np.array([3,-5,4])
v3 = np.array([4,0,-3])

#Orthogonl Vector
dot_product_v1_v2 = np.dot(v1,v2)
print("Dot product of v1 and v2 is :",dot_product_v1_v2)
dot_product_v2_v3 = np.dot(v2,v3)
print("Dot product of v2 and v3 is :",dot_product_v2_v3)
dot_product_v1_v3 = np.dot(v1,v3)
print("Dot product of v1 and v3 is :",dot_product_v1_v3)

#Orthonormal Vectors
print("\n Orthonormal Vectors are :\n")
orthonormal_v1 = v1/np.linalg.norm(v1)
print("V1 =",orthonormal_v1)
orthonormal_v2 = v2/np.linalg.norm(v2)
print("V2 =",orthonormal_v2)
orthonormal_v3 = v3/np.linalg.norm(v3)
print("V3 =",orthonormal_v3)

#Angels between Vectors
print("Angels between Vectors :\n")
angle_v1_v2 = np.arccos(dot_product_v1_v2/(np.linalg.norm(v1)*np.linalg.norm(v2)))
print("Angle between v1 and v2 is :", np.degrees(angle_v1_v2))

angle_v2_v3 = np.arccos(dot_product_v2_v3/(np.linalg.norm(v2)*np.linalg.norm(v3)))
print("Angle between v2 and v3 is :", np.degrees(angle_v2_v3))

angle_v1_v3 = np.arccos(dot_product_v1_v3/(np.linalg.norm(v1)*np.linalg.norm(v3)))
print("Angle between v1 and v3 is :", np.degrees(angle_v1_v3))
```
