```python
# Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Using List Comprehension to get each value of y for each value of x
hours=[1,2,6,7,8,9,11,12]
number_of_cell=[2**x for x in hours ]
print("Different time instance of hours :",hours)
print("Numbers of cells at different time interval :",number_of_cell)
```

```python
# Calculate number of  bins by Sturges' Rule
num_bin= int(np.ceil(1+np.log2(len(hours))))
print("Numbers  of bin :",num_bin)
```

```python
# Plot Histogram in term of relative freqency
plt.hist(number_of_cell, bins=num_bin, density= True, edgecolor= 'black')
plt.title("E-coli Bacteria Cells Distribution by Histogram")
plt.xlabel("Number of E-coli Bacteria Cells")
plt.ylabel("Relative Frequency")
plt.show()
```
